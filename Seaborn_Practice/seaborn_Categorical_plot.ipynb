```python
import seaborn as sns
import numpy as np
```

```python
tips= sns.load_dataset('tips')
tips
```

```python
#stripplot---> values are overrided 
sns.stripplot(x='day',y='total_bill', data=tips,hue='sex')
```

```python
sns.stripplot(x='day',y='total_bill', jitter=False,data=tips,hue='sex')
```

```python
#swarmplot--> no override 
sns.swarmplot(x='day',y='total_bill', data=tips,hue='time')
```

```python
sns.boxplot(x="day",y ="total_bill", data=tips, hue='sex')
```

```python
#violineplot
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.stripplot(x='day',y='total_bill', data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True);
```

```python
sns.barplot(x='day',y='total_bill',data=tips,hue='sex') #default shows Mean
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.sum) 
```

```python
datalable=sns.barplot(x='day',y='total_bill',data=tips,estimator=np.sum) 
for i in datalable.containers:
    datalable.bar_label(i,)
```

```python
#count plot 
sns.countplot(x='day',data=tips)
```

```python
#for datalable
datalable=sns.countplot(x='day',data=tips,hue='sex')
for i in datalable.containers:
    datalable.bar_label(i,)
```

```python
datalable=sns.countplot(x='size',data=tips,hue='sex')
for i in datalable.containers:
    datalable.bar_label(i,)
```
