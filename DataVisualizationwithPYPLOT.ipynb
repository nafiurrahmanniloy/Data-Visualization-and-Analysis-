```python
import matplotlib.pyplot as plt
```

```python
#basic scatter plot
plt.scatter(x,y)
```

```python
plt.plot(x,y,'r-*'); # r-*--> r means red colour, - means solid line,  * as dots.
# ;-----> this hides meta data can use plt.show() for same task

#adding label
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Basic Plot')
plt.xticks(range(0,11));  #ticks to show the values in x or y axis 
plt.yticks(range(1,110,10));
```

```python
#subplot
plt.subplot(1,3,1) # row, column, position 
plt.plot(x,y,'b-*')
plt.subplot(1,3,2) # row, column, position 
plt.plot(y,x,'g-*')
plt.subplot(1,3,3) # row, column, position 
plt.plot(x,y,'r--*')

plt.tight_layout() #output looks better 
```

OOP Concept in matplotlib for vizualization 
Better for size management
Better for ploting and subploting 



```python
fig1 = plt.figure(figsize=(9,5)) #sets a empty place for this fig 
axes = fig1.add_axes([.2,.1,.5,.5]) #boundary for the fig ([left, bottom, width, height])
axes.plot(x,y,'r-*');
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Figure Title')
axes.set_ylim([0,110])
axes.set_xlim([1,11])
fig1.savefig("graphsave.png", dpi=200)  #path 
```

```python
#subplot in OOP
fig, axes= plt.subplots(2,2, figsize=(10,5))  #() to make array give value inside suplot funciton 
plt.tight_layout()
axes[0][0].plot(x,y,'g--o');
axes[0][1].scatter(x,y);
#axes[1][0].
#axes[1][1].
```

```python
# --------------Fig inside fig-----------------------
fig3=plt.figure(figsize=(12,5))
axes_main=fig3.add_axes([.1,.1,1,1])
axes_inside=fig3.add_axes([.2,.5,.3,.4])
#do basic labeling and title
axes_main.plot(x,y,'g');
axes_inside.plot(y,x,'b-o');
fig3.savefig("FigInFig.png", dpi=20)
```

```python
fig, axes = plt.subplots()
axes.plot(x, x+1, 'r.-') # red line with dots
axes.plot(x, x+2, 'g--') # green dashed line
```

```python
fig, axes = plt.subplots()

axes.plot(x, x+1, color="blue", alpha=.1) # half-transparant
axes.plot(x, x+2, color="#8B008B")        # RGB hex code
axes.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

```python
fig, axes = plt.subplots(figsize=(8,6))
axes.plot(x, y, color="purple", lw=3, ls='--', marker='o', markersize=5, markerfacecolor="yellow", 
          markeredgewidth=1, markeredgecolor="red")
#Change parameters e.g. markeredgewidth, markerfacecolor, markeredgecolor, marker etc to see the difference
# let's go the the reference jupyter notebook to see some options!
```

```python
fig, axes = plt.subplots(figsize=(10,8))

axes.plot(x, x+1, color="red", linewidth=0.25)
axes.plot(x, x+2, color="red", linewidth=0.50)
axes.plot(x, x+3, color="red", linewidth=1.00)
axes.plot(x, x+4, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
axes.plot(x, x+5, color="green", lw=5, linestyle='-')
axes.plot(x, x+6, color="green", lw=5, ls='-.')
axes.plot(x, x+7, color="green", lw=5, ls=':')

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
axes.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
axes.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
axes.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
axes.plot(x, x+12, color="blue", lw=3, ls='--', marker='1', markersize=20)

# marker size and color
axes.plot(x, x+13, color="purple", lw=3, ls='-', marker='o', markersize=2)
axes.plot(x, x+14, color="purple", lw=3, ls='-', marker='o', markersize=4)
axes.plot(x, x+15, color="purple", lw=3, ls='-', marker='o', markersize=8, markerfacecolor="red")
axes.plot(x, x+16, color="purple", lw=3, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```
