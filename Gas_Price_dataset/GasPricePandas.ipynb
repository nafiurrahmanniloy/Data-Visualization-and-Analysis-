```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
```

```python
gas = pd.read_csv('gas_prices.csv')
```

```python
gas.describe()
```

```python
gas['UK'].plot()
```

```python
plt.style.use('ggplot')
```

```python
gas.plot(x='Year', y ='UK')
```

```python
#year not showing properly so have to use list 
gas.plot(x='Year', y =['UK','Japan'], xticks=gas.Year[::3].tolist());
```

```python
#doing with pyplot
plt.subplots(figsize=(10,5))
plt.title('Gas Prices over time(IN USD)',fontdict={'fontweight':'bold', 'fontsize': 18, 'color':'g'})
plt.plot(gas['Year'],gas['UK'],label='UK',marker='o',markeredgecolor='red')
plt.plot(gas['Year'],gas['Canada'],label='Canada',marker='o',markeredgecolor='red')
plt.plot(gas['Year'],gas['Germany'],label='Germany',marker='o',markeredgecolor='red')
plt.plot(gas['Year'],gas['Italy'],label='Italy',marker='o',markeredgecolor='red')
plt.xticks(gas.Year[::3].tolist())
plt.legend();
```
