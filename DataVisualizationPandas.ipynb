```python
import pandas as pd 
import numpy as np
```

```python
#periods number of dates to generate 
dates =pd.date_range('1/1/2024',periods= 500)
dates 
```

```python
#randn range (+3 to -3)
data1=np.random.randn(500,3) #shape(row,column) 
data1
```

```python
df1= pd.DataFrame(data= data1,index=dates,columns=['A','B','C'])
```

```python
df1
```

```python
df1['D']=np.random.rand(500) #rand range(0 to 1)
```

```python
df1
```

```python
data2=np.random.rand(20,3)
df2=pd.DataFrame(data=data2,columns=['X','Y','Z'])
```

```python
df2
```

```python
#stacked data
df2.plot(kind='area') #line chart and area chart same
```

```python
#bar plot
df2['X'].plot(kind='bar',figsize=(10,5))
```

```python
df2.plot.bar()
```

```python
#stacked bar
df2.plot(kind='bar',stacked=True,figsize=(10,5))
```

```python
#horizontal  
df2.plot(kind='barh',stacked=True,figsize=(10,5))
```

```python
#scatter plot
df1.plot.scatter(x='A',y='B',c='.8',cmap='coolwarm',figsize=(10,5)); #cmap-->color tone,c is opacity
```

```python
df1.plot.line(x='A',y='B',figsize=(10,5),lw=.3) #same as scatter 
```

```python
#bubble chart---> 3 variable chart 
df1.plot.scatter(x='A',y='B',c='D',cmap='coolwarm',figsize=(10,5)) #this color is dependent on the value of D column 
#s=value---> size of bubbles...can place Other Column here as well
```

```python
#size according to the value of D column 
df1.plot.scatter(x='A',y='B',s=df1['D']*20,c='D',cmap='coolwarm',figsize=(10,5))
```

```python
#hexbin--->shows the density of the data 
df1.plot.hexbin(x='A',y='B',gridsize=50,cmap='Oranges',figsize=(10,5))  
```
